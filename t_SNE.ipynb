```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
```

```python
df=pd.read_csv("/content/TSNE_data.csv")
df.head()
```

```python
df.shape
```

```python
array=df.values
X=array[:,1:]
Y=array[:,0]
```

```python
#TSNE visualization
data_tsne=TSNE(n_components=2).fit_transform(X)
cluster.tsneplot(score=data_tsne)
```

```python
data_tsne
```

```python
newdf=pd.DataFrame(data_tsne).join(df['diagnosis'])
newdf
```

```python
color_class=df['diagnosis'].to_numpy()
cluster.tsneplot(score=data_tsne,colorlist=color_class,legendpos="upper right",legendanchor=(1.15,1))
plt.show()
```
